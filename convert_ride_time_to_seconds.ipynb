```python
import pandas as pd

# Load your file
df = pd.read_csv('rapido_1.0.csv')

def parse_mm_ss(time_str):
    try:
        if pd.isna(time_str):
            return 0
        parts = time_str.split(':')
        if len(parts) == 2:
            minutes = int(parts[0])
            seconds = float(parts[1])
            total_seconds = minutes * 60 + seconds
            return total_seconds
        else:
            return 0
    except:
        return 0

# Apply the function to create the 'duration_seconds' column
df['duration_seconds'] = df['time'].apply(parse_mm_ss)

# Save the updated file
df.to_csv('rapido_cleaned.csv', index=False)

# Preview
print(df[['time', 'duration_seconds']].head())
print(f"Total rows after conversion: {df.shape[0]}")
```
